```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Reading data from CSV file

```python
aditya = pd.read_csv("datasets/Airquality.csv")
aditya.head(12)
```

```python
aditya.isnull().sum()
```

# Data Cleaning

```python
aditya = aditya.drop("Unnamed: 0",axis=1)
aditya.head(6)
```

<h3>Replacing numerical null values</h3>

```python
aditya['Ozone']=aditya['Ozone'].fillna(aditya['Ozone'].mean())
aditya['Solar.R']=aditya['Solar.R'].fillna(aditya['Solar.R'].mean())
aditya["Wind"] = aditya["Wind"].fillna(aditya["Wind"].mean())
```

<h3>Replacing categorical null values</h3>

```python
aditya['Humidity']=aditya['Humidity'].fillna(aditya['Humidity'].mode()[0])
aditya.isnull().sum()
```

```python
aditya.dtypes
```

# Data Transformation
<h3>Using Label Encoding on Humidity column</h3>

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
aditya['Humidity']=label.fit_transform(aditya['Humidity'])
aditya['Humidity'].unique()
```

```python
aditya.dtypes
```

# Visualising the Data
<h2>1. Scatter Plot</h2>

```python
plt.scatter(x = aditya["Ozone"],y = aditya["Temp"])
plt.legend(["Ozone"])
plt.xlabel("Ozone")
plt.ylabel("Temp")
plt.title("Ozone Scatter Plot")
plt.show()
```

<h2>2. Bar Plot</h2>

```python
import seaborn as sns
sns.barplot(aditya["Month"],aditya["Temp"])
```

<h3>3. Heatmap</h3>

```python
corr = aditya.corr()
sns.heatmap(corr, annot = True)
```

<h3>4. Pairplot</h3>

```python
sns.pairplot(aditya)
```

<h2>5. Line Graph</h2>

```python
h = aditya.iloc[1:15, 0]
v = aditya.iloc[1:15, 3]
plt.plot(h, label="Ozone", marker="o", linestyle="dotted")
plt.plot(v, label="Temp", marker="o", linestyle="dashed")
plt.title("Line Graph for Ozone and Temp")
plt.legend()
plt.show()
```

<h2>6. Box Plot</h2>

```python
sns.boxplot(x=aditya["Month"],y=aditya["Temp"])
```

<h2>7. Pie-Chart</h2>

```python
labels= ['Ozone','Solar.R','Wind','Temperature']
sizes=[aditya['Ozone'].mean(),aditya["Solar.R"].mean(),aditya['Wind'].mean(),aditya["Temp"].mean()]
colors=['red','pink','yellow','silver']
textprops =  {"fontsize":15}
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=14, radius=1.5, textprops=textprops)
plt.title("Airquality Factors", fontsize=20, style="italic", pad=35)
```

<h2>8. Histogram</h2>

```python
h=aditya.iloc[:,-1]
plt.hist(h,bins='auto')
plt.title('Histogram ')
plt.xlabel("Humidity")
```
