## Sort Values

### a. Descending Order

````python
desc_sort = aditya.sort_values('share',ascending=False)
desc_sort
````

### b. Ascending Order

````python
asc_sort = aditya.sort_values('share',ascending=True)
asc_sort
````

## Tranpose of Dataset

````python
aditya.transpose()
````
